```python
#import data using the lust yiu had generated
crash_df = sns.load_dataset('car_crashes')
crash_df
```

DISTRIBUTION PLOTS

```python
#write seaborn code to draw a distribution plot
sns.displot(crash_df["not_distracted"],kde = False,bins =25)
```

JOINT PLOT

```python
#plot a joint plot
sns.jointplot(x = "speeding", y = "alcohol",data= crash_df,kind="reg")
```

```python
#plot a joint plot
sns.jointplot(x = "speeding", y = "alcohol",data= crash_df,kind="kde")
```

```python
#plot a joint plot
sns.jointplot(x = "speeding", y = "alcohol",data= crash_df,kind="hex")
```

KDE PLOT

```python



sns.kdeplot(crash_df["alcohol"],label = "alcohol")
plt.legend()
```

PAIR PLOTS

```python
sns.pairplot(crash_df)
```

```python
tips_df = sns.load_dataset("tips")
sns.pairplot(tips_df,hue="sex",palette="Blues")
```

RUG PLOT

```python
#plot a rugplot the rug plot to visualize the dataset
sns.rugplot(tips_df["tip"])
```
