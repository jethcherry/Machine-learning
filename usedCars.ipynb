```python

cat_cols = user_data.select_dtypes(include=['object']).columns
num_cols= user_data.select_dtypes(include = np.number).columns.to_list()
print("categorical Variables")
print(cat_cols)
print("Numerical Variables")
print(num_cols)
```

```python


for col in  num_cols:
    print(col)
    print('Skew :', round(user_data[col].skew(),2))
    plt.figure(figsize=(15,4))
    plt.subplot(1,2,1)
    user_data[col].hist(grid=False)
    plt.ylabel('count')
    plt.subplot (1,2,2)
    sns.boxplot(x= user_data[col])
     
```

```python
fig, axes = plt.subplots(3, 2, figsize=(18, 18))
fig.suptitle('Bar plot for all categorical variables in the dataset')

sns.countplot(ax=axes[0, 0], x='Fuel_Type', data=user_data, color='blue', order=user_data['Fuel_Type'].value_counts().index)
sns.countplot(ax=axes[0, 1], x='Transmission', data=user_data, color='blue', order=user_data['Transmission'].value_counts().index)
sns.countplot(ax=axes[1, 0], x='Owner_Type', data=user_data, color='blue', order=user_data['Owner_Type'].value_counts().index)
sns.countplot(ax=axes[1, 1], x='Location', data=user_data, color='blue', order=user_data['Location'].value_counts().index)
sns.countplot(ax=axes[2, 0], x='Brand', data=user_data, color='blue', order=user_data['Brand'].value_counts().head(20).index)
sns.countplot(ax=axes[2, 1], x='Model', data=user_data, color='blue', order=user_data['Model'].value_counts().head(20).index)

axes[1, 1].tick_params(axis='x', labelrotation=45)
axes[2, 0].tick_params(axis='x', labelrotation=90)
axes[2, 1].tick_params(axis='x', labelrotation=90)

plt.tight_layout()
plt.show()
```

```python
#Function for log transformation of the column
def log_transform(user_data, col) :
    for colname in col:
        if (user_data[colname] == 1.0).all():
        
            user_data[colname + '_log'] = np.log(user_data[colname ]+1)
           
        else:
         user_data[colname + '_log'] = np.log(user_data[colname])
user_data.info()

log_transform(user_data,['Kilometers_Driven','Price'])

#log_transformation of the feature 'Kilometers_Driven'
sns.displot(user_data["Kilometers_Driven_log"], axlabel="Kilometers_Driven_log")
        
```
