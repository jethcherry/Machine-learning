```python
#import data using the lust yiu had generated
crash_df = sns.load_dataset('car_crashes')
crash_df
```

DISTRIBUTION PLOTS

```python
#write seaborn code to draw a distribution plot
sns.displot(crash_df["not_distracted"],kde = False,bins =25)
```

JOINT PLOT

```python
#plot a joint plot
sns.jointplot(x = "speeding", y = "alcohol",data= crash_df,kind="reg")
```

```python
#plot a joint plot
sns.jointplot(x = "speeding", y = "alcohol",data= crash_df,kind="kde")
```

```python
#plot a joint plot
sns.jointplot(x = "speeding", y = "alcohol",data= crash_df,kind="hex")
```

KDE PLOT

```python



sns.kdeplot(crash_df["alcohol"],label = "alcohol")
plt.legend()
```

PAIR PLOTS

```python
sns.pairplot(crash_df)
```

```python
tips_df = sns.load_dataset("tips")
sns.pairplot(tips_df,hue="sex",palette="Blues")
```

```python
print(tips_df["day"])
```

RUG PLOT

```python
#plot a rugplot the rug plot to visualize the dataset
sns.rugplot(tips_df["tip"])
```

STYLING

```python
sns.set_style("ticks")
plt.figure(figsize=(8,4))
sns.set_context("paper",font_scale=1.4)
sns.despine(left="False")
sns.jointplot(x = "speeding",y="alcohol",data= crash_df,kind = "reg")
```

CATEGORICAL PLOTS

BAR PLOTS

```python
sns.barplot(x="sex",y="total_bill",data = tips_df,estimator=np.var)
```

COUNT PLOTS

```python
#use count to count the number of occurences
sns.countplot(x = "sex",data = tips_df)
```

BOX PLOT

```python
#plotting a boxplot
sns.boxplot(x="day",y="total_bill",data = tips_df,hue="sex")
plt.legend(loc = 0)
```

VIOLIN PLOT

```python
sns.violinplot(x="day",y="total_bill",data = tips_df,hue="sex",split="True")
plt.legend(loc=0)
```

```python
sns.violinplot(x = "day",y="total_bill",data = tips_df)
```

STRIP PLOTS

```python
plt.figure(figsize=(8,5))
sns.stripplot(x = "day",y="total_bill",data = tips_df,jitter= True,hue="sex",dodge=True)
plt.legend(loc=0)
```

SWARM PLOT

```python

sns.swarmplot(x = "day",y="total_bill",data = tips_df,color="white")
```
