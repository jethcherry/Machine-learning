```python
#Draw four subplots in the same figure and setting their titles

x = np.arange(100)

fig,axs = plt.subplots(2,2)

axs[0,0].plot(x,np.sin(x))
axs[0,0].set_title("Sine Wave")

axs[0,1].plot(x,np.cos(x))
axs[0,1].set_title("Cosine Wave")

axs[1,0].plot(x,np.random.random(100))
axs[1,0].set_title("Random Function")

axs[1,1].plot(x,np.log(x))
axs[1,1].set_title("Log Function")

#removes the overlap between the headings and the label
plt.tight_layout()

axs[1,1].set_xlabel("TEST")
fig.suptitle("Four plots")

#saves the subplot in a png picture format
plt.savefig("fourplots.png", dpi = 900, transparent = False,pad_inches = 20.2)



```

```python

#create  3d figure using  scatter plots
ax = plt.axes(projection = "3d")

x = np.random.random(100)

y = np.random.random(100)

z = np.random.random(100)

ax.scatter(x,y,z)
plt.tight_layout()
ax.set_title("3D Plot Visualization")
ax.set_xlabel("test")
```

```python
#create simalar 3d figure using  line graph
ax = plt.axes(projection = "3d")

x = np.arange(0,50,0.1)

y = np.arange(0,50,0.1)

z = np.cos(x+y)

ax.plot(x,y,z)
plt.tight_layout()
ax.set_title("3D Plot Visualization")
ax.set_xlabel("test")
```

```python
#create simalar 3d figure but showing surfaces or a surface plot
ax = plt.axes(projection = "3d")

x = np.arange(-5,5,0.1)
y = np.arange(-5,5,0.1)

X,Y = np.meshgrid(x,y)

Z = np.sin(X) * np.cos(Y)


ax.plot_surface(X,Y,Z)

plt.tight_layout()
ax.set_title("3D Plot Visualization")
ax.set_xlabel("test")
```

```python
#create simalar 3d figure but showing surfaces or a surface plot
ax = plt.axes(projection = "3d")

x = np.arange(-5,5,0.1)
y = np.arange(-5,5,0.1)

X,Y = np.meshgrid(x,y)

Z = np.sin(X) * np.cos(Y)


ax.plot_surface(X,Y,Z,)
ax.view_init(azim = 0,elev = 90)
plt.tight_layout()
ax.set_title("3D Plot Visualization")
ax.set_xlabel("test")
```

```python
#create simalar 3d figure but showing surfaces or a surface plot
ax = plt.axes(projection = "3d")

x = np.arange(-5,5,0.1)
y = np.arange(-5,5,0.1)

X,Y = np.meshgrid(x,y)

Z = np.sin(X) * np.cos(Y)


ax.plot_surface(X,Y,Z,cmap = "Spectral")
ax.view_init(azim = 0,elev = 90)
plt.tight_layout()
ax.set_title("3D Plot Visualization")
ax.set_xlabel("test")
```
